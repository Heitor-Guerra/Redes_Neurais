```python
import torch
```

## Exemplo Simples


```python

a = torch.tensor(3.0, requires_grad=True)
b = torch.tensor(3.0, requires_grad=True)

y = a ** 2 + b
z = torch.sin(y)

# caminha para trás no grafo computacional
# calculando as derivadas parciais de z em
# relação aos nós
z.backward()

print(a.grad)
print(b.grad)
```

## Gradient Descent

```python
x = torch.tensor(5.0, dtype=torch.float32, requires_grad=True)

for _ in range(10):
    # computa funcao objetivo
    f = x ** 2

    # calcula a derivada parcial da funcao
    # objetivo em relacao ao x
    f.backward()

    # faz um passo de descida do gradiente
    with torch.no_grad():
        x -= 0.3 * x.grad

    x.grad.zero_()
    # veja como x vai tender ao minimo (0, no caso de x**2)
    print(x.item())
```
